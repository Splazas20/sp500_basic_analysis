```python
# Paso 1: Configuración del Entorno de Trabajo

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA

# Paso 2: Cargar y Preprocesar los Datos

# Darle acceso a que Colab consulte Drive
from google.colab import drive
drive.mount('/content/drive')

# Subir el archivo de csv a Colab
file_path = '/content/drive/My Drive/Estudio/(2024-02) Tech Bootcamp/Actividades estudiantes - Análisis de datos/JUAN SEBASTIÁN PLAZAS/Componente técnico/Proyecto - Etapa 5/S&P500_prices_proccessed.csv'
df = pd.read_csv(file_path)

# Confirmar si hay nulos antes de elminarlos
print("\nPrimeras filas del Dataframe subido por CSV a Colab:\n\n",df.head())
print("\nCantidad de null dentro de los datos:\n\n",df.isnull().sum())
print("\nInformación del df antes de quitar null:\n")
print(df.info())

# Eliminar los nulos y verificar cómo quedó el df
df_sin_nulos = df.dropna()
print("\nVerificación de la cantidad de null luego de quitarlos:\n\n",df_sin_nulos.isnull().sum())

# Confirmar que la columna Date quede en formato datetime
print("\n Tipo de dato de Date antes de convertirlo: ",df['Date'].dtype)
df['Date'] = pd.to_datetime(df['Date'])
print("\n Tipo de dato de Date después de convertirlo: ",df['Date'].dtype)

# Ordenar por Symbol y Date
df_ordenado = df.sort_values(by=['Symbol', 'Date'])
print("\nPrimeras filas del Dataframe ordenado:\n\n",df_ordenado.head())

# Paso 3: Cálculo de Retornos Porcentuales Diarios

# Agrupar el DataFrame por 'Symbol' y calcular la variación porcentual diaria para cada grupo
grupos = df_ordenado.groupby('Symbol')
df_variacion = grupos['Close'].pct_change()
print("\nPrimeras filas del Dataframe variación porcentual:\n\n",df_variacion.head())

# Agregar la variación porcentual al DataFrame ordenado
df_ordenado['Return'] = df_variacion
print("\nPrimeras filas del Dataframe con variación porcentual:\n\n",df_ordenado.head())

# Eliminar los nulos que surgieron de return y verificar cómo quedó el df
print("\nVerificación de la cantidad de null antes de quitarlos:\n\n",df_ordenado.isnull().sum())
df_return_sin_null = df_ordenado.dropna()
print("\nVerificación de la cantidad de null luego de quitarlos:\n\n",df_return_sin_null.isnull().sum())

# Paso 4: Cálculo de Indicadores de Volatilidad

# Agrupar el DataFrame por 'Symbol' y calcular la desviación de Return para cada grupo
grupos = df_return_sin_null.groupby('Symbol')
df_desviacion = grupos['Return'].std()
print("\nPrimeras filas del Dataframe con desviacion:\n\n",df_desviacion.head())

# Calcular la diferencia entre el máximo y mínimo
df_diferencia = grupos['Return'].max() - grupos['Return'].min()
print("\nPrimeras filas del Dataframe con la diferencia entre el máximo y el mínimo:\n\n",df_diferencia.head())

# Calcular la media absoluta de 'Return' para cada grupo
df_mad_por_symbol = grupos['Return'].apply(lambda x: np.mean(np.abs(x - x.mean())))
print("\nPrimeras filas del Dataframe con la media absoluta:\n\n",df_mad_por_symbol.head())

# Unir el dataframe de desviación, diferencia y MAD
df_resultados = pd.DataFrame({
    'Desviacion': df_desviacion,
    'Max-Min': df_diferencia,
    'MAD': df_mad_por_symbol
})
print("\nPrimeras filas del Dataframe con los resultados:\n\n",df_resultados.head())

# Paso 5: Escalamiento de los Datos

# Crear un nuevo DataFrame para los datos escalados
df_estandarizada = df_return_sin_null.copy()
df_normalizada = df_return_sin_null.copy()

# Para la estandarización, creamos un objeto StandardScaler y lo aplicamos a las columnas Close y Return
scaler = StandardScaler()
df_estandarizada[['Close', 'Return']] = scaler.fit_transform(df_estandarizada[['Close', 'Return']])
print("\nPrimeras filas del Dataframe estandarizado:\n\n",df_estandarizada.head())

# Para la normalización, creamos un objeto MinMaxScaler y lo aplicamos a las columnas Close y Return
scaler = MinMaxScaler()
df_normalizada[['Close', 'Return']] = scaler.fit_transform(df_normalizada[['Close', 'Return']])
print("\nPrimeras filas del Dataframe normalizado:\n\n",df_normalizada.head())

# Paso 6: Clusterización

# Definir rangos y arreglo para los métodos
datos = []
rango_codo = range(1, 6)
rango_silueta = range(2, 6)

# Usando los datos estandarizados

# Método de codo

# Calcular la inercia para cada número de clusters
for k in rango_codo:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_estandarizada[['Close', 'Return']])
    datos.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(rango_codo, datos, 'bx-')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.title('Método del Codo para Encontrar el Número Óptimo de Clusters con datos estandarizados')
plt.show()

# Método de la silueta

# Calcular la puntuación de la silueta para cada número de clusters
datos = []

for k in rango_silueta:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(df_estandarizada[['Close', 'Return']])
    score = silhouette_score(df_estandarizada[['Close', 'Return']], labels)
    datos.append(score)

plt.figure(figsize=(10, 6))
plt.plot(rango_silueta, datos, 'bx-')
plt.xlabel('Número de Clusters')
plt.ylabel('Puntuación de la Silueta')
plt.title('Método de la Silueta para Encontrar el Número Óptimo de Clusters con datos estandarizados')
plt.show()

# Usando los datos normalizados

# Método de codo

# Calcular la inercia para cada número de clusters
datos = []

for k in rango_codo:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_normalizada[['Close', 'Return']])
    datos.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(rango_codo, datos, 'bx-')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.title('Método del Codo para Encontrar el Número Óptimo de Clusters con datos normalizados')
plt.show()

# Método de la silueta

# Calcular la puntuación de la silueta para cada número de clusters
datos = []

for k in rango_silueta:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(df_normalizada[['Close', 'Return']])
    score = silhouette_score(df_normalizada[['Close', 'Return']], labels)
    datos.append(score)

plt.figure(figsize=(10, 6))
plt.plot(rango_silueta, datos, 'bx-')
plt.xlabel('Número de Clusters')
plt.ylabel('Puntuación de la Silueta')
plt.title('Método de la Silueta para Encontrar el Número Óptimo de Clusters con datos estandarizados')
plt.show()

# Conclusión para usar 2 clusters
print("\nEstudiando los métodos normalizados y estandarizados con los métodos de codo y silueta, voy a usar 2 clusters.")
print("\nEn los análisis del método codo, aunque no se muy claro, la disminución de la inercia se ve en el 2 o 3.")
print("\nEn los análisis del método silueta, claramente hay una tendencia en el cluster 2")



```

```python
# Uso de K-means para agrupar las empresas según las métricas de volatilidad

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(df_estandarizada[['Close', 'Return']])

# Agregar las etiquetas al DataFrame con datos estasndarizados
df_estandarizada['Cluster'] = kmeans.labels_
print("\nPrimeras filas del Dataframe con las etiquetas de los clusters:\n\n",df_estandarizada.head())

df_estandarizada

# Paso 8: Visualización de los Resultados

plt.scatter(df_estandarizada['Close'], df_estandarizada['Return'], c=df_estandarizada['Cluster'], cmap='viridis')
plt.title('Agrupación de Empresas por Volatilidad')
plt.xlabel('Close Estandarizado')
plt.ylabel('Return Estandarizado')
plt.colorbar(label='Cluster')
plt.show()

df_con_datos = df_return_sin_null.copy()
df_con_datos['Cluster'] = df_estandarizada['Cluster']
df_con_datos.head()

plt.scatter(df_con_datos[df_con_datos['Cluster'] == 0]['Close'], df_con_datos[df_con_datos['Cluster'] == 0]['Return'], label='Cluster 0')
plt.scatter(df_con_datos[df_con_datos['Cluster'] == 1]['Close'], df_con_datos[df_con_datos['Cluster'] == 1]['Return'], label='Cluster 1')
plt.title('Agrupación de Empresas por Volatilidad')
plt.xlabel('Close')
plt.ylabel('Return')
plt.legend()
plt.show()
```
